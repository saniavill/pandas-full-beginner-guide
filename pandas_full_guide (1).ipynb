We can select different sections of our dataframe. in the following code we're selecting column 'movie' along with column 'rating' to display each movie and its rating.

There is a reason we're using 2 brackets in the cell above:<br>Everytime we're selecting more than one column to display, we have to put the columns' names inside a list and then inside df[]. Double brackets return a DataFrame while single brackets for one column return a Series.

#### Before we dive into data manipulation, we should know where to learn about our DataFrame.

```python
df.info()
```

#### The info() function gives us information about the datatype of the values inside each column.
for example in our movie dataframe, the column 'rating', has a float64 datatype, we can change that data type for less memory usage. There are two ways to do that:
1. Change all of our columns' datatypes to float32. This would work on dataframes in which all the columns include floats/ints but in our case we also have object datatypes so this method would cause an error.
2. Select the columns that are of type float64, and convert those columns:

```python
float64_cols = list(df.select_dtypes(include='float64'))
df[float64_cols] = df[float64_cols].astype('float32')
```

We use the list() and select_dtypes() functions to create a list of all the data that are of type float64, we will then select that list in the next line and use the astype() function to change the datatypes to float32.

```python
df.info()
```

As you can see, our 'rating' column is now of type float32 that results in our dataframe having a slightly less memory usage.

Back to displaying certain parts of your DataFrame, one efficient way to do that is by using pandas' loc() and iloc() functions <br> Let's say we want to display the first 2 rows and the first 2 columns only. Here's how we would do that using the loc() function:

```python
df.loc[:1, ['movie', 'genre']]
```

Notice that while slicing in loc(), the end index is *inclusive* <br>
* NOTE: the loc() function in pandas is primarily label-based, meaning it works with column names (labels) rather than numerical indices. Therefore, you can't directly slice columns by their numerical indices using loc. However, as you can see, you can slice rows.
If you need to select columns by their position or by slicing, you should use iloc. Let's display the first 5 rows and all the columns from 'director' to 'genre' (that have the indices of 1 and 3):

```python
df.iloc[:5, 1:4]
```

As you can see in the code above, the end index in iloc() is exclusive. And you can use it to slice both columns and rows. But you can't use columns' labels to display them.<br>To summarize:
* Use loc() for selecting columns by their names.
* Use iloc() for slicing columns by integer positions.

Let's say we want to replace the rating values of the first 4 rows to NaN:

```python
import numpy as np
df_copy = df.copy()
df_copy.loc[:3, ['rating']] = np.nan
df_copy.head()
```

Now to re-fill NaN values, we use the fillna() method:

```python
df_copy = df_copy.fillna(9)
df_copy.head()
```

Let's get back to our original dataframe and get all the movies in which the rating in between 6 and 8 (8 being exclusive):

```python
df_new = df.copy()
df_new[df_new['rating'].between(6, 8)]
```

Now how about we get all the movies in which the director is Christopher Nolan and the rating is higher than 8:

```python
df_new = df.copy()
df_new[(df_new['director'] == 'Christopher Nolan') & (df_new['rating'] > 8)]
```
